```python
from autogluon.tabular import TabularDataset, TabularPredictor
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

train.shape, test.shape
```

```python
train.drop(['id'], axis=1, inplace=True)
test.drop(['id'], axis=1, inplace=True)
train.shape, test.shape
```

```python
train_data = TabularDataset(train)
test_data = TabularDataset(test)

predictor = TabularPredictor(problem_type='binary',label='target',eval_metric='f1').fit(train_data=train_data,presets='best_quality',ag_args_fit={'num_gpus': 1})
predictions = predictor.predict(test_data)
```

```python
predictor.leaderboard()
```

```python
y_pred = predictor.predict(test_data)
y_pred
```

```python
submission = pd.DataFrame({'id':sample['id'],'target':y_pred.tolist()})
submission.to_csv('samp_autoG.csv',index=False)
submission.head()
```
