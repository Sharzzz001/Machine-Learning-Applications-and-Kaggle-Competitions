```python
rem_list = ['hsc_p','etest_p','AVG_GPA_CT','AVG_GPA_MANG']
```

```python
#Using the InterQuartile Range to fill the values
def remove_outlier(i , col):
    Q1 = i[col].quantile(0.25)
    Q3 = i[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_whisker = Q1-1.5*IQR
    upper_whisker = Q3+1.5*IQR
    df1[col] = np.clip(i[col] , lower_whisker , upper_whisker)
    return i

def treat_outliers_all(df , col_list):
    for c in col_list:
        i = remove_outlier(df , c)
    return i
```

```python
df2 = treat_outliers_all(df2 , rem_list)
```

lgbm

```python
import lightgbm as lgb
import xgboost as xgb
import optuna
import optuna.integration.lightgbm as lgbo


class config:
    OPTUNA_SWITCH = "on"
    OBJECTIVE = "regression"
    METRIC = "rmse"
    TEST_SIZE = 0.2
```

```python
from sklearn.preprocessing import MinMaxScaler

cols = X_train.columns
scaler = MinMaxScaler()
X_train_RF = scaler.fit_transform(X_train)
X_test_RF = scaler.transform(X_test)

X_train_RF = pd.DataFrame(X_train_RF,columns=cols)
X_test_RF = pd.DataFrame(X_test_RF,columns=cols)
```

```python
def tune_parameters_lgbm(X_train_RF , y_train):
    opt_params = {
        "objective" : config.OBJECTIVE,
        "metric" : config.METRIC
    }

    reg_train = lgb.Dataset(
        X_train_RF , 
        y_train
    )
    
    reg_eval = lgb.Dataset(
        X_test_RF , 
        y_test,
        reference = reg_train
    )
    
    opt = lgbo.train(
        opt_params,
        reg_train,
        valid_sets = reg_eval ,
        verbose_eval = False,
        num_boost_round = 200,
        early_stopping_rounds = 100
    )
    
    return opt
if config.OPTUNA_SWITCH == "on":
    opt = tune_parameters_lgbm(X_train_RF , y_train)  
```

```python
print("Best Parameter For LGBM is")
print(opt.params)
```

```python
lgbm_params = {'objective': 'regression', 'metric': 'rmse', 'feature_pre_filter': False, 
               'lambda_l1': 3.907267756456269e-07, 'lambda_l2': 8.323761819617874e-08, 
               'num_leaves': 31, 'feature_fraction': 0.652, 'bagging_fraction': 1.0, 'bagging_freq': 0, 
               'min_child_samples': 10, 'num_iterations': 200, 'early_stopping_round': 100}

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=config.TEST_SIZE, random_state=0)

X_train_RF = scaler.fit_transform(X_train)
X_test_RF = scaler.transform(X_test)

X_train_RF = pd.DataFrame(X_train_RF,columns=cols)
X_test_RF = pd.DataFrame(X_test_RF,columns=cols)


lgb_train = lgb.Dataset(X_train_RF, y_train)
lgb_eval = lgb.Dataset(X_test_RF, y_test, reference=lgb_train)
evaluation_results = {}                       
model_lgbm = lgb.train(
    lgbm_params,
    valid_names=['train', 'valid'],           
    valid_sets=[lgb_train, lgb_eval],        
    evals_result=evaluation_results,          
    train_set=lgb_train
)
```

```python
y_pred_lgbm = model_lgbm.predict(X_test_RF)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
from math import sqrt

mae_lgbm = mean_absolute_error(y_test,y_pred_lgbm)
rmse_lgbm = sqrt(mean_squared_error(y_test,y_pred_lgbm))
r_sq_lgbm = r2_score(y_test,y_pred_lgbm)

print("Mean Absolute Error: {0}\nRoot Mean Square Error: {1}\nR2 Score: {2}".format(round(mae_lgbm,3),round(rmse_lgbm,3),round(r_sq_lgbm,3)))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
# can change the range to get better results
def objective(trial):
    # Define the hyperparameters to optimize
    param = {
        'objective': 'reg:squarederror',
        'eval_metric': 'rmse',
        'reg_lambda': trial.suggest_float('reg_lambda', 1e-3, 1e2, log=True),
        'learning_rate': trial.suggest_float('learning_rate', 1e-4, 1e0, log=True),
        'n_estimators': trial.suggest_int('n_estimators', 50, 200, step=15),
        'lambda': trial.suggest_float('lambda', 1e-8, 1.0),
        'alpha': trial.suggest_float('alpha', 1e-8, 1.0),
        'max_depth': trial.suggest_int('max_depth', 3, 9),
        'eta': trial.suggest_float('eta', 1e-8, 1.0),
        'gamma': trial.suggest_float('gamma', 1e-8, 1.0),
        'subsample': trial.suggest_float('subsample', 0.1, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.1, 1.0),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 300),
        'n_jobs': -1
    }

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2, random_state=0)

    X_train_RF = scaler.fit_transform(X_train)
    X_test_RF = scaler.transform(X_test)

    X_train_RF = pd.DataFrame(X_train_RF,columns=cols)
    X_test_RF = pd.DataFrame(X_test_RF,columns=cols)

    # Train the model
    model = xgb.XGBRegressor(**param)
    model.fit(X_train_RF, y_train)

    # Evaluate the model
    y_pred = model.predict(X_test_RF)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    return rmse
```

```python
# Can change the number of trials to check for better results
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=300)
```

```python
print("Important parameters For XGB is")
print(optuna.importance.get_param_importances(study))
```

```python
best_params = study.best_trial.params
print("Best Parameter For XGB is")
print(best_params)
```

```python
model_xgb = xgb.XGBRegressor(objective='reg:squarederror', **best_params)
model_xgb.fit(X_train,y_train)
```

```python
from catboost import CatBoostRegressor

def objective(trial):
    X_train, X_test, y_train, y_test = model_selection.train_test_split(X,y, test_size=0.2,random_state=0)

    X_train_RF = scaler.fit_transform(X_train)
    X_test_RF = scaler.transform(X_test)

    X_train_RF = pd.DataFrame(X_train_RF,columns=cols)
    X_test_RF = pd.DataFrame(X_test_RF,columns=cols)

    param = {
        "objective": trial.suggest_categorical("objective", ["RMSE","MAE"]),
        "colsample_bylevel": trial.suggest_float("colsample_bylevel", 0.01, 0.1),
        "depth": trial.suggest_int("depth", 1, 12),
        "learning_rate": trial.suggest_float("learning_rate",0.01,0.2),
        "boosting_type": trial.suggest_categorical("boosting_type", ["Ordered", "Plain"]),
        "bootstrap_type": trial.suggest_categorical(
        "bootstrap_type", ["Bayesian", "Bernoulli", "MVS"]
        ),
        "used_ram_limit": "3gb",
    }

    if param["bootstrap_type"] == "Bayesian":
        param["bagging_temperature"] = trial.suggest_float("bagging_temperature", 0, 10)
    elif param["bootstrap_type"] == "Bernoulli":
        param["subsample"] = trial.suggest_float("subsample", 0.1, 1)

    gbm = CatBoostRegressor(**param)

    gbm.fit(X_train_RF, y_train, eval_set=[(X_test_RF, y_test)], verbose=0, early_stopping_rounds=100)

    # Evaluate the model
    y_pred = gbm.predict(X_test_RF)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    return rmse
```

```python
study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=100, timeout=600)
```

```python
print("Important parameters For cat is")
print(optuna.importance.get_param_importances(study))
```

```python
best_params_cat = study.best_trial.params
print("Best Parameter For cat is")
print(best_params_cat)
```

```python
model_cat = CatBoostRegressor(**best_params_cat)
model_cat.fit(X_train_RF,y_train)
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler
from sklearn.pipeline import Pipeline
# Importing libraries for building the neural network
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score, KFold, train_test_split
from keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau, TensorBoard
from keras.optimizers import SGD, Adam, Adadelta, RMSprop
from keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import tensorflow as tf
```

```python
placement_model = Sequential()
placement_model.add(Dense(256, input_shape = (X_train_RF.shape[1],), activation = "relu"))
placement_model.add(Dense(512, activation = "relu"))
placement_model.add(Dense(1024, activation = "relu"))
placement_model.add(Dropout(0.2))
placement_model.add(Dense(1024, activation = "relu"))
placement_model.add(Dense(512, activation = "relu"))
placement_model.add(Dense(256, activation = "relu"))
placement_model.add(Flatten())
placement_model.add(Dense(512, activation = "relu"))
placement_model.add(Dropout(0.5))
placement_model.add(Dense(512, activation = "relu"))
placement_model.add(Dropout(0.5))
placement_model.add(Dense(1, activation = "linear",kernel_regularizer = tf.keras.regularizers.l1(l=0.001)))
placement_model.summary()
```

```python
optimizer=Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999)

# Compile the network :
placement_model.compile(optimizer=optimizer, loss=tf.keras.losses.MAE, metrics=[tf.keras.metrics.MAE])
```

```python
m_loss = EarlyStopping(
    monitor="val_loss",
    min_delta=0,
    patience=500,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=True,
)
```

```python
save_checkpoint = tf.keras.callbacks.ModelCheckpoint(filepath='rent_model.h5', monitor='val_loss', save_best_only=True, verbose=1)
```

```python
history=placement_model.fit(X_train_RF, y_train, epochs=1000,validation_split=0.2,batch_size=1, verbose=1,callbacks=[m_loss,save_checkpoint])
```

```python
dfh = pd.DataFrame(history.history)
plt.figure(figsize=(10, 6))
plt.title('Engg Salary Prediction Model History')
plt.plot(dfh)
plt.xlabel('epochs')
plt.legend(dfh.columns)
plt.show()
```
