```python
from autogluon.tabular import TabularDataset, TabularPredictor
import pandas as pd

train_data = pd.read_csv('Placement_Train.csv')
test_data = pd.read_csv('Placement_Test.csv')

train_data.shape, test_data.shape
```

```python
train_data.info()
```

```python
train_data.drop('SR_no',axis=1, inplace=True)
test_data.drop('SR_no',axis=1, inplace=True)
```

```python
train_data = TabularDataset(train_data)
test_data = TabularDataset(test_data)

predictor = TabularPredictor(problem_type='regression',label='Annual_salary',eval_metric='root_mean_squared_error').fit(train_data=train_data,presets='best_quality',ag_args_fit={'num_gpus': 1})
predictions = predictor.predict(test_data)
```

```python
predictor.leaderboard()
```

```python
y_pred = predictor.predict(test_data)
y_pred
```

```python
sample = pd.read_csv('Plcement_Sample_Submission.csv')
# sample.head()
submission = pd.DataFrame({'SR_no':sample['SR_no'],'Annual_salary':y_pred})
submission.to_csv('samp.csv',index=False)
submission.head()
```
